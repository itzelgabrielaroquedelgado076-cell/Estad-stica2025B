```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/itzelgabrielaroquedelgado076-cell/Estad-stica2025B/refs/heads/main/Habitos_Deportivos.csv")
df

```

```python
df["Deporte Extracurricular"]=df["Deporte Extracurricular"].map({"Yes":1,"No":0})
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

matriz_de_correlacion= df.corr()

plt.figure(figsize=(10,5), dpi= 100)

sns.heatmap(
    matriz_de_correlacion, #lo que va a graficar
    annot= True, #se muestran los valores en las celdas
    fmt= "0.2f", #0.2f significa "dos decimales"
    cmap= "coolwarm", #paleta de colores
    square= True, #forzar cuadrados
    linewidths= 0.5, #grosor de lineas
    cbar_kws= {"shrink":0.8}, #tamaño barra lateral
    center= 0,
    )

plt.title("Mapa de color para matriz de correlacion", fontsize=20)
plt.xticks(rotation= 45, ha= "right")
plt.yticks(rotation= 0)

```

```python
# 1. Establesca una variable dependiente ( Y ) y una variable independiente ( X ).
import pandas as pd


df["Entrenamientos_Semanales"] = df["Entrenamientos Semanales"]
df["Horas_de_Entrenamiento"] = df["Horas de Entrenamiento"]


x = df["Entrenamientos_Semanales"]
y = df["Horas_de_Entrenamiento"]

# 2. Realice un diagrama de dispersión para estos datos.
import matplotlib.pyplot as plt
plt.figure(figsize = (16, 8), dpi = 100)
plt.scatter(x, y, label = "Entrenamientos Semanales", color = 'red')
plt.xlabel('Entrenamientos Semanales') # nombrar eje x
plt.ylabel('Horas de Entrenamiento') # nombrar eje y
plt.gca().spines[["top","right"]].set_visible(False)


# 3. ¿Los datos soportan la suposición de linealidad?
# Sí soportan la susposición de linealidad ya que aunque los puntos no
#parezcan estar en forma de linea imaginaria representan una tendencia positiva y recta pero con una
#gran parte de variabilidad

# 4. Calcule el coeficiente de correlación e interprete el resultado.
from scipy.stats import pearsonr
r, _ = pearsonr(x, y)
print(f'Coeficiente de correlación: {r: 0.4f}\n')
#Si hay correlación, pero de una manera moderada ya que aunque los entrenamientos semanales pueden aumentar, pero las horas de entrenamiento
#no aumentan de la misma manera proporcional.

# 6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de
# dispersión.
import statsmodels.api as sm
x_constante = sm.add_constant(x)
modelo = sm.OLS(y, x_constante).fit()

yc = modelo.predict(x_constante)

plt.plot(x, yc, color = 'black', linestyle = '--')

# 5. Calcule el coeficiente de determinación e interprete el resultado.
from sklearn.metrics import r2_score  # recomendada
r2 = r2_score(y, yc)
print(f'Coeficiente de determinación: {r2: 0.4f}\n')
#El 24.3% de la variación en las horas de entrenamiento se pueden explicar por los entrenamientos semanales
#mientras que el restante de 75.7% se debe a otros factores, lo que no permite confirmar que la relación es lineal
#pero débil o moderada ya que no se relacionan tanto.

# 7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de
# regresión ajustada ( b1 )
nivel_de_confianza = 0.95
intervalo_de_confianza = modelo.conf_int(alpha = 1 - nivel_de_confianza)
intervalo_de_confianza_b1 = intervalo_de_confianza.iloc[1]
print(f'Intervalo de confianza para b1 de {nivel_de_confianza: 0.0%}')
print(f'{intervalo_de_confianza_b1[0]: 0.4f} < b1 < {intervalo_de_confianza_b1[1]: 0.4f}\n')
#Los valores obtenidos con el intervalo de confianza del 95% nos indican que si hay correlación ya que en ambos valores
#no se incluye el cero, lo que nos permite confirmar la correlación o una relación significativa y esto nos dice al igual que los entrenamientos
#semanales tienen un efecto positivo en las horas de entrenamiento y aunque estos pueden variar lo que se espera es que
#cada entrenamiento adicional incremente las horas entre 0.06(4 minutos mínimo) y 0.93(56 minutos de incremento máximo) horas de entrenamiento.


# 8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente,
# ¿Parece que se verifican los supuestos?
residuales = modelo.resid
plt.figure()
plt.scatter(x, residuales, color = 'blue')
plt.xlabel('Entrenamientos Semanales')
plt.ylabel('Horas de Entrenamiento')
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
#Conforme al gráfico de residuales si parece que se verfican los supuestos ya que los errores o residuos
#se encuentran de manera aleatoria arriba y abajo del cero


# 9. Realice la prueba de Shapiro para los residuales y comente el resultado.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f'valor-p de Shapiro: {valor_p_sh: 0.4f}\n')
#Conforme la prueba de shapiro se puede confirmar que hay normalidad ya que el valor-p
#es mayor al nivel de significancia por lo que se acepta la Ho:Normalidad

# 10. Realice la prueba de Breusch-Pagan para los residuales y comente el
# resultado.
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')
#Por medio de la prueba de Breusch podemos confirmar que si hay homoscedasticidad debido a que el
#valor-p es mayor al nivel de significancia por lo que no se rechaza la Ho:La varianza de los errores es constante

# 11.
print("x=0",modelo.predict([[1,0]]))
print("x=4",modelo.predict([[1,4]]))
#Para x=0 las horas de entrenamiento estimadas son igual a 3.9 casi 4 horas y para x=4 las horas de entrenamiento estimadas
#son de casi 6 horas por lo que se puede corfirmar un poco la tendencia creciente moderada, ya que hay un crecimiento no tan alto o constante
#en el que las horas de entrenamiento aumentan de una manera moderada cuando aumentan los entrenamientos semanales.

#12
from statsmodels.formula.api import ols
modelo_lineal = ols('Horas_de_Entrenamiento~Entrenamientos_Semanales', data=df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
#La tabla ANOVA nos indica que los entrenamientos semanales tienen un efecto estadísticamente
#significativo en las horas de entrenamiento lo que nos indica que los entrenamientos semanales explican una parte de la
# variabilidad en las horas lo que al igual se puede observar en el valor-p que es menor al nivel de significancia y al ser así
#se confirma lo anteriormente mencionado.

#Conclusión final
#Conforme a cada uno de los cálculos realizados anteriormente podemos confirmar que el modelo es confiable ya que se cumplieron cada uno de los supuestos
#lo que nos indica que el número de entrenamientos semanales es un buen predictor lo que valida el modelo ajustado y nos permite interpretar
#los resultados con confianza donde las recomendaciones que se extren son de que cada entrenamiento semanal de una manera adcional se asocia con un
#incremento promedio obtenido en nuestro intervalo de confianza ya que existe una relación positiva  y significativa aunque moderada entre cada una de las variables, en donde por cada entrenamiento
#adicional se esperea un incremento promedio de horas que esta entre 4 y 56 minutos
```
